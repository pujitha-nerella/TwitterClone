```python
import numpy as np
import torch
from transformers import BertTokenizer, BertModel
from sklearn.metrics.pairwise import cosine_similarity

# Function to calculate accuracy using BERT embeddings and cosine similarity
def calculate_bert_accuracy(questions, passages):
    accuracies = []

    # Load pre-trained BERT tokenizer and model
    tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
    model = BertModel.from_pretrained('bert-base-uncased')

    for i in range(len(questions)):
        question = questions[i]
        passage = passages[i]

        # Tokenize question and passage
        input_ids = tokenizer.encode(question, passage, add_special_tokens=True, return_tensors='pt')

        # Get BERT embeddings
        with torch.no_grad():
            outputs = model(input_ids)
            question_embedding = outputs.last_hidden_state[0, 0].numpy()
            passage_embedding = outputs.last_hidden_state[0, 1].numpy()

        # Calculate cosine similarity
        similarity = cosine_similarity([question_embedding], [passage_embedding])[0][0]

        # Use cosine similarity as accuracy (normalized to [0, 1] range)
        accuracy = (similarity + 1) / 2  # Normalize to [0, 1]
        accuracies.append(accuracy)

    return accuracies

# Sample questions and passages
questions = [
    "Who is the author of the Mistborn series?",
    "What are the names of the three books in the first series?",
    "Where does the series take place?",
    "How many powers does a Misting possess?"
]
passages = [
    "Mistborn is a series of epic fantasy novels written by American author Brandon Sanderson.",
    "The first series, published between 2006 and 2008, consists of The Final Empire, The Well of Ascension, and The Hero of Ages.",
    "The series primarily takes place in a region called the Final Empire on a world called Scadrial, where the sun and sky are red, vegetation is brown, and the ground is constantly being covered under black volcanic ashfalls.",
    '"Mistings" have only one of the many Allomantic powers, while "Mistborns" have all the powers.'
]

# Calculate accuracies using BERT embeddings and cosine similarity
bert_accuracies = calculate_bert_accuracy(questions, passages)

# Calculate the overall accuracy for BERT
bert_overall_accuracy = sum(bert_accuracies) / len(bert_accuracies)

# Print the overall accuracy for BERT
print("BERT Overall Accuracy:", bert_overall_accuracy)
```

```python
###sequence to sequence model
import json
import numpy as np
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.models import Model
from keras.layers import Input, LSTM, Dense, Embedding

# Load data from JSON files
with open("train.json", "r") as train_file:
    train_data = json.load(train_file)

with open("test.json", "r") as test_file:
    test_data = json.load(test_file)

# Process train data
train_texts = []
train_labels = []
for item in train_data:
    train_texts.append(item['context'])
    train_labels.append(item['qas'][0]['question'])

# Process test data
test_texts = []
test_labels = []
for item in test_data:
    test_texts.append(item['context'])
    test_labels.append(item['qas'][0]['question'])

# Tokenize texts
tokenizer = Tokenizer()
tokenizer.fit_on_texts(train_texts + train_labels)

# Convert texts to sequences
train_question_sequences = tokenizer.texts_to_sequences(train_labels)
train_response_sequences = tokenizer.texts_to_sequences(train_texts)

test_question_sequences = tokenizer.texts_to_sequences(test_labels)
test_response_sequences = tokenizer.texts_to_sequences(test_texts)

# Pad sequences to ensure uniform length
max_seq_length = max(len(seq) for seq in train_question_sequences + test_question_sequences)
train_question_padded = pad_sequences(train_question_sequences, maxlen=max_seq_length, padding='post')
train_response_padded = pad_sequences(train_response_sequences, maxlen=max_seq_length, padding='post')

test_question_padded = pad_sequences(test_question_sequences, maxlen=max_seq_length, padding='post')
test_response_padded = pad_sequences(test_response_sequences, maxlen=max_seq_length, padding='post')

# Define vocabulary size
vocab_size = len(tokenizer.word_index) + 1

# Define Seq2Seq model architecture
embedding_dim = 100
latent_dim = 256

# Encoder
encoder_inputs = Input(shape=(max_seq_length,))
encoder_embedding = Embedding(input_dim=vocab_size, output_dim=embedding_dim)(encoder_inputs)
encoder_lstm = LSTM(latent_dim, return_state=True)
encoder_outputs, state_h, state_c = encoder_lstm(encoder_embedding)
encoder_states = [state_h, state_c]

# Decoder
decoder_inputs = Input(shape=(max_seq_length,))
decoder_embedding = Embedding(input_dim=vocab_size, output_dim=embedding_dim)(decoder_inputs)
decoder_lstm = LSTM(latent_dim, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(decoder_embedding, initial_state=encoder_states)
decoder_dense = Dense(vocab_size, activation='softmax')
decoder_outputs = decoder_dense(decoder_outputs)

# Define the model
model = Model([encoder_inputs, decoder_inputs], decoder_outputs)

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit([train_question_padded, train_response_padded], train_response_padded,
                    epochs=5, batch_size=64, validation_split=0.2)

# Evaluate the model
test_loss, test_accuracy = model.evaluate([test_question_padded, test_response_padded], test_response_padded)
print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)
```

```python
import json
import numpy as np
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.models import Model
from keras.layers import Input, LSTM, Dense, Embedding
from keras.utils import plot_model

# Load data from JSON files
with open("train.json", "r") as train_file:
    train_data = json.load(train_file)

with open("test.json", "r") as test_file:
    test_data = json.load(test_file)

# Process train data
train_texts = []
train_labels = []
for item in train_data:
    train_texts.append(item['context'])
    train_labels.append(item['qas'][0]['question'])

# Process test data
test_texts = []
test_labels = []
for item in test_data:
    test_texts.append(item['context'])
    test_labels.append(item['qas'][0]['question'])

# Tokenize texts
tokenizer = Tokenizer()
tokenizer.fit_on_texts(train_texts + train_labels)

# Convert texts to sequences
train_question_sequences = tokenizer.texts_to_sequences(train_labels)
train_response_sequences = tokenizer.texts_to_sequences(train_texts)

test_question_sequences = tokenizer.texts_to_sequences(test_labels)
test_response_sequences = tokenizer.texts_to_sequences(test_texts)

# Pad sequences to ensure uniform length
max_seq_length = max(len(seq) for seq in train_question_sequences + test_question_sequences)
train_question_padded = pad_sequences(train_question_sequences, maxlen=max_seq_length, padding='post')
train_response_padded = pad_sequences(train_response_sequences, maxlen=max_seq_length, padding='post')

test_question_padded = pad_sequences(test_question_sequences, maxlen=max_seq_length, padding='post')
test_response_padded = pad_sequences(test_response_sequences, maxlen=max_seq_length, padding='post')

# Define vocabulary size
vocab_size = len(tokenizer.word_index) + 1

# Define Seq2Seq model architecture
embedding_dim = 100
latent_dim = 256

# Encoder
encoder_inputs = Input(shape=(max_seq_length,))
encoder_embedding = Embedding(input_dim=vocab_size, output_dim=embedding_dim)(encoder_inputs)
encoder_lstm = LSTM(latent_dim, return_state=True)
encoder_outputs, state_h, state_c = encoder_lstm(encoder_embedding)
encoder_states = [state_h, state_c]

# Decoder
decoder_inputs = Input(shape=(max_seq_length,))
decoder_embedding = Embedding(input_dim=vocab_size, output_dim=embedding_dim)(decoder_inputs)
decoder_lstm = LSTM(latent_dim, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(decoder_embedding, initial_state=encoder_states)
decoder_dense = Dense(vocab_size, activation='softmax')
decoder_outputs = decoder_dense(decoder_outputs)

# Define the model
model = Model([encoder_inputs, decoder_inputs], decoder_outputs)

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit([train_question_padded, train_response_padded], train_response_padded,
                    epochs=5, batch_size=64, validation_split=0.2)

# Evaluate the model
test_loss, test_accuracy = model.evaluate([test_question_padded, test_response_padded], test_response_padded)
print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)

# Visualize the model architecture and save it as an image
plot_model(model, to_file='seq2seq_model.png', show_shapes=True, show_layer_names=True)
```
